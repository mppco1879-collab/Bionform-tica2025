#### 6.- Realiza una función que realice una digestión con las siguientes enzimas de restricción en nuestra secuencia. La función tiene que devolver la siguiente información: (2 pto)
1. Enzimas que tienen sitios de corte en la secuencia y posición (o posiciones de corte)
2. Enzimas que no tienen sitios de corte (si las hay)
3. Resultado de la digestión: los fragmentos resultantes de nuestra secuencia
4. Indicar número y tamaño de cada fragmento

```python
enzR = {
    'HindIII' : 'A*AGCTT',
    'BamHI' : 'G*GATCC',
    'AluI' : 'AG*CT',
    'Sau3AI' : '*GATC',
    'EcoRI' : 'G*AATTC'
}
```

```python
#he tenido que meter el codigo junto y sin muchas lineas porque no 
#me permitia el añadir mas lineas por algún bug raro 

#el codigo lee la secuencia de ADN, busca 
import re
from Bio import SeqIO
#dicciionario con enzimas de restriccion 
enzR = {
    'HindIII': 'A*AGCTT',
    'BamHI': 'G*GATCC',
    'AluI': 'AG*CT',
    'Sau3AI': '*GATC',
    'EcoRI': 'G*AATTC'
}

def digestion_simple(fasta_file, enzimas):
    # para le lectura, abre el archivo FASTA y se obtiene la secuencia 
    # de ADN, upper convierte todos los nucleotidos a mayusculas 
    record = SeqIO.read(fasta_file, "fasta")
    seq = str(record.seq).upper()

    enzimas_con_corte = {}
    enzimas_sin_corte = []

    # Se buscan los sitios de corte, re.finditer busca todas las apariciones del sitio 
    #de corte dentro del ADN 
    #m.start y sitio.index, calcula la posición exacta donde la enzima corta 
    for enzima, sitio in enzimas.items():
        patron = sitio.replace("*", "")  
        posiciones = [m.start() + sitio.index("*") for m in re.finditer(patron, seq)]
#si hay posiciones encontradas se guarda en enzimas_con_corte, si no hay, se 
#se añade el nombre a enzimas_sin_corte 
        if posiciones:
            enzimas_con_corte[enzima] = posiciones
        else:
            enzimas_sin_corte.append(enzima)

    # En caso de que haya cortes, se generan fragmentos
    #une todas las posicions de corte en uns sola lsita y las ordena,
    #recorre toda la secuencia cortandola en esos puntos y guarda cada
    #fragmento entre cortes seguidos en la lista de fragmentos
    cortes = sorted([p for pos in enzimas_con_corte.values() for p in pos])
    fragmentos = []
    inicio = 0
    for fin in cortes:
        fragmentos.append(seq[inicio:fin])
        inicio = fin
    fragmentos.append(seq[inicio:])

    # para los resultados, muestra enzimas que si cortaron y en que posiciones del ADN
    print(" Enzimas con corte:")
    for e, p in enzimas_con_corte.items():
        print(f"  {e}: posiciones {p}")
#Muestra las enzimas que no tienen sitio de corte en la secuencia
    print("\n Enzimas sin corte:")
    print("  " + ", ".join(enzimas_sin_corte) if enzimas_sin_corte else "  Ninguna")
#Imprime cada fragmento generado, indicando su numero y tamaño en pb
    print("\n Fragmentos resultantes:")
    for i, f in enumerate(fragmentos):
        print(f"  Fragmento {i+1}: {len(f)} pb")

    print(f"\nNúmero total de fragmentos: {len(fragmentos)}")

    return enzimas_con_corte, enzimas_sin_corte, fragmentos 
    
```

```python
fasta_file = "CNTNAP2.fasta"
enzimas_con_corte, enzimas_sin_corte, fragmentos = digestion_simple(fasta_file, enzR)
```

#### 7.- Realiza una función que mediante la introducción de una secuencia de ADN devuelva lo siguiente: (3pto)
1. Todos los ORFs posibles que puedan surgir de la secuencia.
2. Selecciona los 6 ORFs más largos e indica si corresponden a la strand positiva o negativa.
3. Devuelve como mejor predicción, el más largo de todos.
4. Calcula el porcentaje de aminoácidos aromáticos de esa proteína elegida y su longitud relativa a la secuencia original de ADN.

```python
from Bio import SeqIO
from Bio.Seq import Seq

def orf_estudiante(fasta_file):
    # Leer secuencia y definimos una funcion llamada orf_estudiante. 
    #Analiza la secuencia de ADN y los ORFs 
    #Se abre el archivo FASTA y guarda la secuencia en record, se convierte
    #todo a mayusculas y el resultado se guarda en la variable dna 
    record = SeqIO.read(fasta_file, "fasta")
    dna = record.seq.upper()
#Aqui se encuentran los ORFs encontrados, junto con otra informacion
    orfs = []

    # Hebra positiva, se recorre los frames y se traduce la secuencia en aminoacidos
    #sin detenerse en el primer codon STOP, al traducirlo, los codones STOP se marcan con *
    #se usa .split para separar la proteina en fragmentos entre los codones stop 
    for frame in range(3):
        seq = dna[frame:]
        prot = seq.translate(to_stop=False)
        pedazos = str(prot).split('*')  # * indica codón stop
        for p in pedazos: #recorre fragmentos y busca si hay M que representa el codon de inicio ATG 
            if 'M' in p:  # M = codón de inicio (ATG)
                start = p.index('M')
                orfs.append({'prot': p[start:], 'strand': '+', 'frame': frame+1})

    # Hebra negativa, calcula la hebra complementaria inversa 
    rev = dna.reverse_complement()
    for frame in range(3):
        seq = rev[frame:]
        prot = seq.translate(to_stop=False)
        pedazos = str(prot).split('*')
        for p in pedazos:
            if 'M' in p:
                start = p.index('M')
                orfs.append({'prot': p[start:], 'strand': '-', 'frame': frame+1})

    # Ordenar por longitud, se usa la funcion lambda para calcular la longitud de la proteina 
    #reverse= True hace que el orden sea descendente 
    orfs.sort(key=lambda x: len(x['prot']), reverse=True)

    # Mostrar los 6 más largos, para cada uno indica, en que hebra se encuentra, en que marco de lectura y su longtud 
    print("Los 6 ORFs más largos:\n")
    for i, o in enumerate(orfs[:6]):
        print(f"ORF {i+1}: Hebra {o['strand']} | Frame {o['frame']} | Longitud {len(o['prot'])} aa")

    # El mas largo, con posicion 0, se considera la mejor prediccion del gen 
    #Se guarda su secuencia de aminoacidos en best_prot
    best = orfs[0]
    best_prot = best['prot']

    #Cuenta cuantos aa aromaticos hay 
    #Cacula su porcentaje con respecto al total en la proteina 
    arom = best_prot.count('F') + best_prot.count('W') + best_prot.count('Y')
    porc_arom = (arom / len(best_prot)) * 100 if len(best_prot) > 0 else 0

    #Cada aa proviene de 3 nucleotidos 
    #Se multiplica la long de la proteina por 3 y divide por long total del ADN
    #Develve el % de ADN que ocupa el mejor ORF  
    rel = (len(best_prot)*3 / len(dna)) * 100

    print("\nMejor ORF:")
    print(f"Hebra: {best['strand']} | Frame: {best['frame']}")
    print(f"Longitud: {len(best_prot)} aa")
    print(f"% Aromáticos: {porc_arom:.2f}%")
    print(f"Longitud relativa al ADN: {rel:.2f}%")

    return best
```
